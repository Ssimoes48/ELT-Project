```python
import pandas as pd
```

```python
happy_15 =  "../Resources/happy_2015.csv"
happy_16 =  "../Resources/happy_2016.csv"
happy_17 =  "../Resources/happy_2017.csv"
happy_18 =  "../Resources/happy_2018.csv"
happy_19 =  "../Resources/happy_2019.csv"

happy_15_df = pd.read_csv(happy_15, encoding="utf-8")
happy_16_df = pd.read_csv(happy_16, encoding="utf-8")
happy_17_df = pd.read_csv(happy_17, encoding="utf-8")
happy_18_df = pd.read_csv(happy_18, encoding="utf-8")
happy_19_df = pd.read_csv(happy_19, encoding="utf-8")
```

```python

happy_15_df["Year"] = "2015"

h15 = happy_15_df.rename(columns={"Happiness Rank": "Overall rank", "Country" : "Country or region", 
                                  "Happiness Score": "Score", "Economy (GDP per Capita)": "GDP per capita", 
                                  "Family": "Social support", "Health (Life Expectancy)": "Healthy life expectancy", 
                                  "Freedom": "Freedom to make life choices", 
                                  "Trust (Government Corruption)": "Perceptions of corruption"})

h15_clean = h15[["Overall rank", "Country or region", "Score", "GDP per capita", "Social support", 
                 "Healthy life expectancy", "Freedom to make life choices", "Generosity", 
                 "Perceptions of corruption", "Year"]]
h15_clean.head()
```

```python
happy_16_df["Year"] = "2016"

h16 = happy_16_df.rename(columns={"Happiness Rank": "Overall rank", "Country" : "Country or region", 
                                  "Happiness Score": "Score", "Economy (GDP per Capita)": "GDP per capita", 
                                  "Family": "Social support", "Health (Life Expectancy)": "Healthy life expectancy", 
                                  "Freedom": "Freedom to make life choices", 
                                  "Trust (Government Corruption)": "Perceptions of corruption"})

h16_clean = h16[["Overall rank", "Country or region", "Score", "GDP per capita", "Social support", 
                 "Healthy life expectancy", "Freedom to make life choices", "Generosity", 
                 "Perceptions of corruption", "Year"]]
h16_clean.head()
```

```python
happy_17_df["Year"] = "2017"

h17 = happy_17_df.rename(columns={"Happiness.Rank": "Overall rank", "Country" : "Country or region", 
                                  "Happiness.Score": "Score", "Economy..GDP.per.Capita.": "GDP per capita", 
                                  "Family": "Social support", "Health..Life.Expectancy.": "Healthy life expectancy", 
                                  "Freedom": "Freedom to make life choices", 
                                  "Trust..Government.Corruption.": "Perceptions of corruption"})

h17_clean = h17[["Overall rank", "Country or region", "Score", "GDP per capita", "Social support", 
                 "Healthy life expectancy", "Freedom to make life choices", "Generosity", 
                 "Perceptions of corruption", "Year"]]
h17_clean.head()
```

```python
happy_18_df["Year"] = "2018"

h18_clean = happy_18_df[["Overall rank", "Country or region", "Score", "GDP per capita", "Social support", 
                 "Healthy life expectancy", "Freedom to make life choices", "Generosity", 
                 "Perceptions of corruption", "Year"]]
h18_clean.head()
```

```python
happy_19_df["Year"] = "2019"

h19_clean = happy_19_df[["Overall rank", "Country or region", "Score", "GDP per capita", "Social support", 
                 "Healthy life expectancy", "Freedom to make life choices", "Generosity", 
                 "Perceptions of corruption", "Year"]]
h19_clean.head()
```

```python

happy_clean = pd.concat([h15_clean,h16_clean,h17_clean,h18_clean,h19_clean,])

happy_clean.head()

happy_clean_new = happy_clean.rename(columns={"Overall rank": "happiness_rank", "Country or region": "country", 
                                  "Score": "happiness_score", "GDP per capita": "gdp_per_capita", 
                                  "Social support": "social_support", "Healthy life expectancy": "healthy_life_expectancy", 
                                              "Freedom to make life choices": "choice_freedom", "Generosity": "generosity", 
                                              "Perceptions of corruption": "perceptions_of_corruption", "Year": "year"})
happy_clean_new.head()
```

```python
# The path to our CSV file
file = "../Resources/countries.csv"

countries_df = pd.read_csv(file, encoding='UTF-8')
new_countries_df = countries_df["country"]

new_countries_df.head()
```

```python
merge_df = pd.merge(new_countries_df, happy_clean_new, on=('country'), how='inner')

merge_df.head()
```

```python
from sqlalchemy import create_engine
from config import postgres_pwd, postgres_user

db_name = "happiness_db"
table_name = "happiness_index"

connection_string = f"{postgres_user}:{postgres_pwd}@localhost:5432/{db_name}"
engine = create_engine(f'postgresql://{connection_string}')
conn = engine.connect()

merge_df.to_sql(table_name, con=engine, if_exists="append", index=False)

conn.close()
```
